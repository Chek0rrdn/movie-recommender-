# 01 - Carga y Limpieza de Datos
## Sistema de Recomendación de Películas basado en Letterboxd

```python
import pandas as pd
import numpy as np
import os
from dotenv import load_dotenv
```

```python
load_dotenv('../.env')
TMDB_API_KEY = os.getenv('TMDB_API_KEY')
print("API Key cargada:", "Sí" if TMDB_API_KEY else "No")
```

### 1.1 Cargar el archivo CSV de Letterboxd

```python
df = pd.read_csv('../data/raw/ratings.csv')
print(f"Total de películas calificadas: {len(df)}")
print(f"\nColumnas: {df.columns.tolist()}")
df.head(10)
```

### 1.2 Exploración inicial

```python
print("Forma del DataFrame:", df.shape)
print("\nTipos de datos:")
print(df.dtypes)
print("\nEstadísticas de calificaciones:")
print(df['Rating'].describe())
print("\nDistribución de calificaciones:")
print(df['Rating'].value_counts().sort_index())
```

### 1.3 Limpieza de datos
Renombramos columnas y eliminamos filas sin calificación.

```python
df_clean = df.rename(columns={
    'Name': 'title',
    'Year': 'year',
    'Rating': 'my_score',
    'Letterboxd URI': 'letterboxd_url',
    'Date': 'date_rated'
})

df_clean = df_clean[df_clean['my_score'] > 0].copy()
df_clean.reset_index(drop=True, inplace=True)
print(f"Películas calificadas (score > 0): {len(df_clean)}")
df_clean.head(10)
```

### 1.4 Exportar datos limpios

```python
df_clean.to_csv('../data/processed/movies_clean.csv', index=False)
print("CSV guardado en data/processed/movies_clean.csv")
```
